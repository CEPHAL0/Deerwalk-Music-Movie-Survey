```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Univariate Analysis


### 1. Distribution of gender


```python
sns.set(style="white")

gender_palette = {"Male": "black", "Female": "pink"}

plt.figure(figsize=(12, 8))
ax = sns.countplot(data=df, x="gender", palette=gender_palette)

for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.0f}",
        (p.get_x() + p.get_width() / 2.0, p.get_height()),
        ha="center",
        va="center",
        xytext=(0, 10),
        textcoords="offset points",
        color="black",
        fontsize=12,
    )

plt.title("Distribution of Gender", fontsize=16)
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Number of Students", fontsize=14)

ax.grid(False)
sns.despine()
plt.show()
```

```python
bivariate_columns_dictionary = {
    "music_streaming_platform": "Music Streaming Platform",
    "language_music_primary": "Primary Language for Music",
    "music_language_secondary": "Secondary Language for Music",
    "fav_music_genre": "Favorite Primary Music Genre",
    "fav_music_genre_2": "Favorite Secondary Music Genre",
    "fav_musical_ins": "Favorite Musical Instrument",
    "play_instrument": "Instrument Played",
    "fav_movie_genre": "Favorite Movie Genre",
    "fav_tvseries_genre": "Favorite TV Series Genre",
    "movie_preference": "Movie Industry Preference",
}

bivariate_columns_rotate_dictionary = {
    "music_streaming_platform": 0,
    "language_music_primary": 0,
    "music_language_secondary": 0,
    "fav_music_genre": 90,
    "fav_music_genre_2": 90,
    "fav_musical_ins": 90,
    "play_instrument": 90,
    "fav_movie_genre": 90,
    "fav_tvseries_genre": 90,
    "movie_preference": 0,
}


gender_palette = {"Male": "black", "Female": "pink"}

for column, description in bivariate_columns_dictionary.items():
    plt.figure(figsize=(12, 6))

    order = df[column].value_counts().index

    ax = sns.countplot(
        x=column,
        hue="gender",
        data=df,
        palette=gender_palette,
        order=order,
    )

    for p in ax.patches:
        if p.get_height() != 0:  # Check if the height is not equal to 0
            ax.annotate(
                f"{p.get_height():.0f}",
                (p.get_x() + p.get_width() / 2.0, p.get_height()),
                ha="center",
                va="center",
                xytext=(0, 10),
                textcoords="offset points",
                color="black",
                fontsize=8,
            )

    plt.xlabel(description)
    plt.ylabel("Number of Students")
    plt.title(f"Distribution of Gender and {description}", fontsize=16)
    plt.xticks(rotation=bivariate_columns_rotate_dictionary[column])
    plt.legend(title="Gender")

    plt.show()
```
